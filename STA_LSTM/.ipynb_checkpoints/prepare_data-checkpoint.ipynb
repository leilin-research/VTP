# Produce back-current-front veh pair

```python
# import prediction results
dsgroup = pd.read_csv('/home/lei/workspace/data/trajectory/dsID.csv')
dsgroup = dsgroup.drop(dsgroup.columns[0], axis = 1)
dsgroup.columns = ['dsid']
```

```python
# import prediction results
veh_id = pd.read_csv('/home/lei/workspace/data/trajectory/vehid.csv')
veh_id = veh_id.drop(veh_id.columns[0], axis = 1)
veh_id.columns = ['vehid']
```

```python
veh_id.head()
```

```python
T = pd.read_csv('/home/lei/workspace/data/trajectory/T.csv')
T = T.drop(T.columns[0], axis = 1)
T.columns = ['current_time']
```

```python
T.head()
```

```python
pred_x = pd.read_csv('/home/lei/workspace/data/trajectory/pred_x.csv')
```

```python
pred_y = pd.read_csv('/home/lei/workspace/data/trajectory/pred_y.csv')
```

```python
T.shape
```

```python
pred_x.shape
```

```python
pred_y.shape
```

```python
pred_x = pred_x.transpose()
pred_x.head()
```

```python
# drop the first row
pred_x = pred_x.iloc[1:].reset_index(drop = True)
pred_x.head()
```

```python
pred_x.columns = ['x_step1','x_step2','x_step3','x_step4','x_step5' ]
```

```python
pred_y = pred_y.transpose()
pred_y.head()
```

```python
# drop the first row
pred_y = pred_y.iloc[1:].reset_index(drop = True)
pred_y.columns = ['y_step1','y_step2','y_step3','y_step4','y_step5' ]
```

```python
prediction_res = pd.concat([dsgroup, veh_id, T, pred_x, pred_y], axis = 1)
prediction_res.head()
```

```python
i = 12 # 2328 some vehicles have repeated time steps e.g., 8453
group = df_ts_group['dataset'][i]
veh_back = df_ts_group['back'][i]
veh_center = df_ts_group['center'][i]
veh_front = df_ts_group['front'][i]

trajectory_bac = prediction_res[(prediction_res['dsid'] == group) & (prediction_res['vehid'] == veh_back)].sort_values(by = 'current_time').reset_index(drop = True)
trajectory_bac.columns = ['dsid', 'backvehid', 'current_time', 'back_x_step1', 'back_x_step2', 'back_x_step3',
       'back_x_step4', 'back_x_step5', 'back_y_step1', 'back_y_step2', 'back_y_step3', 'back_y_step4',
       'back_y_step5']

back_traj = df_ts[(df_ts.iloc[:, 0] == group) & (df_ts.iloc[:, 1] == veh_back)].iloc[:, [1, 2, 3, 4]].reset_index(drop = True)
back_traj.columns = ['backvehid', 'current_time', 'current_back_x', 'current_back_y']
new_traj_back = pd.merge(trajectory_bac, back_traj, how = 'left', on = ['backvehid', 'current_time'])
#new_traj_back.to_csv('backveh_' + str(veh_back)+'_prediction'+'.csv')

trajectory_fro = prediction_res[(prediction_res['dsid'] == group) & (prediction_res['vehid'] == veh_front)].sort_values(by = 'current_time').reset_index(drop = True)
trajectory_fro.columns = ['dsid', 'frontvehid', 'current_time', 'front_x_step1', 'front_x_step2', 'front_x_step3',
       'front_x_step4', 'front_x_step5', 'front_y_step1', 'front_y_step2', 'front_y_step3', 'front_y_step4',
       'front_y_step5']

front_traj = df_ts[(df_ts.iloc[:, 0] == group) & (df_ts.iloc[:, 1] == veh_front)].iloc[:, [1, 2, 3, 4]].reset_index(drop = True)
front_traj.columns = ['frontvehid', 'current_time', 'current_front_x', 'current_front_y']
new_traj_front = pd.merge(trajectory_fro, front_traj, how = 'left', on = ['frontvehid', 'current_time'])
#new_traj_front.to_csv('frontveh_' + str(veh_front)+'_prediction'+'.csv')

# 1-center veh, 2 time frame, 3 center x, 4 center y
center_traj = df_ts[(df_ts.iloc[:, 0] == group) & (df_ts.iloc[:, 1] == veh_center) & (df_ts.iloc[:, 26] == veh_back) & (df_ts.iloc[:, 28] == veh_front)].iloc[:, [1, 2, 3, 4, 26, 28]].reset_index(drop = True)
center_traj.columns = ['cen_vehid', 'current_time', 'cen_x', 'cen_y', 'back_vehid', 'front_vehid']

#center_traj.to_csv('cenveh_traj_'+str(veh_center)+'.csv')

```

```python
center_traj = pd.merge(center_traj, new_traj_back, how = 'left', on = 'current_time')
center_traj = pd.merge(center_traj, new_traj_front, how = 'left', on = 'current_time')
center_traj.to_csv('trajectory_'+'cen_' + str(veh_center)+'_back_'+str(veh_back)+'_front_'+str(veh_front)+'.csv')
```

# Compare the control performance to the real trajectory

```python
res_2414 = pd.read_csv('trajectory_cen_2342.0_back_2387.0_front_2328.0.csv')
```

```python
res_2414.head()
```

```python
res_2414.columns
```

```python
# no negative distance allowed
res_2414.loc[res_2414['back_y_step1']<0, 'back_y_step1'] = 0
res_2414.loc[res_2414['front_y_step1']<0, 'front_y_step1'] = 0
```

```python
# let's control the center vehicle to be in the center of the front and back vehicle 
# based on future trajector prediction

# y is the horizontal distance

backpos = res_2414['current_back_y'] + res_2414['back_y_step1']
frontpos = res_2414['current_front_y'] + res_2414['front_y_step1']

center_veh = backpos + (frontpos - backpos) / 2 
center_veh
```

```python
center_veh = center_veh.values
center_veh = center_veh[0:387] # predict next step (t+1) while it is t, the last time step is useless for compare
```

```python
# compare whether the back veh will hit the center
current_back_y = res_2414['current_back_y'].values[1:] # from time 1 to the end
dist_back_cen = center_veh - current_back_y
print ('number of negative dist (back veh hit center veh if control like this way:', len(dist_back_cen[dist_back_cen < 0]))
```

```python
# compare whether the back veh will hit the center
current_front_y = res_2414['current_front_y'].values[1:] # from time 1 to the end
dist_cen_front = current_front_y - center_veh
print ('number of negative dist (center veh hit front veh if control like this way:', len(dist_cen_front[dist_cen_front < 0]))
```

```python
initial_cen_veh = res_2414['cen_y'].values[0] # initial position of center veh
real_cen_veh = res_2414['cen_y'].values[1:] # real veh trajectory
con_cen_veh = center_veh # controlled veh trajectory
```

```python
# compare speed
real_speed = real_cen_veh[1:] - real_cen_veh[0:-1]
control_speed = con_cen_veh[1:] - con_cen_veh[0:-1]
control_speed
```
